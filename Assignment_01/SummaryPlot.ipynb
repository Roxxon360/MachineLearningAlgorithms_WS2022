```python
import matplotlib.pyplot as plt
import numpy as np
  
test_size = [0.1,0.3,0.5,0.7,0.9]
Gaussian = [57.33,56.28,55.7,56.12,58.94]
Multinomial = [82.83,82.67,82.17,82.26,81.7]
Complement = [70.17,70.22,70.63,70.95,70.48]
Bernoulli = [84.17,83.83,82.63,81.95,81.0]
Categorical = [67.17,62.56,57.3,48.74,30.48]
DecisionTree = [81.33,76.17,73.07,71.29,62.78]

plt.xlabel("test_size")
plt.ylabel("Correctly Labeled [%]")
plt.plot(test_size, Gaussian, label = "Gaussian")
plt.plot(test_size, Multinomial, label = "Multinomial")
plt.plot(test_size, Complement, label = "Complement")
plt.plot(test_size, Bernoulli, label = "Bernoulli")
plt.plot(test_size, Categorical, label = "Categorical")
plt.plot(test_size, DecisionTree, label = "Decision Tree")
plt.legend(loc='center', bbox_to_anchor=(0.5, 1.1),ncol=3, fancybox=True)
plt.rcParams["figure.figsize"] = (12,8)
plt.show()
```
